```python
import requests
from bs4 import BeautifulSoup
import string

def fetch_words_from_url(url):
    """Функция для извлечения слов с заданного URL."""
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        words = []
        for link in soup.find_all("a", href=True):
            href = link["href"]
            if href.startswith("/wiki/") and ":" not in href:
                word = link.text.strip()
                if word:
                    words.append(word)
        return list(set(words))  # Удаляем дубликаты
    else:
        print(f"Не удалось получить страницу {url}. Код ответа: {response.status_code}")
        return []

def save_words_to_file(words, filename):
    """Функция для записи слов в файл."""
    with open(filename, "a", encoding="utf-8") as file:
        for word in sorted(words):  # Сортируем для удобства
            file.write(word + "\n")

# Основной код
base_url = "https://ru.wiktionary.org/wiki/Индекс:Русский_язык/"
output_file = "russian_words_all.txt"
alphabet = [chr(i) for i in range(ord('А'), ord('Я') + 1)]  # Генерируем список букв алфавита

for letter in alphabet:
    url = base_url + letter
    print(f"Обрабатываем URL: {url}")
    words = fetch_words_from_url(url)
    if words:
        save_words_to_file(words, output_file)
        print(f"Добавлено {len(words)} слов из страницы {letter}")

print(f"Слова из всех страниц сохранены в файл {output_file}")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# Функция для извлечения транскрипции с страницы Викисловаря
def get_transcription(word):
    url = f"https://ru.wiktionary.org/wiki/{word}"
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Не удалось получить данные для {word}")
        return None

    soup = BeautifulSoup(response.text, 'html.parser')

    # Ищем транскрипцию
    transcription = None
    for span in soup.find_all("span", {"class": "IPA"}):
        transcription = span.get_text()
        break  # Возьмем только первую найденную транскрипцию

    return transcription

# Чтение слов из файла ru_50k.txt и ограничение до 5000 слов
words = []
with open('russian_words_all.txt', 'r', encoding='utf-8') as file:
    for i, line in enumerate(file.readlines()):
        word = line.split()[0]  # Извлекаем только слово, оставляем частоту
        words.append(word)
        if i >= 10000:  # Ограничиваем до х слов
            break

# Открываем CSV файл для записи
with open('words.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Слово", "Транскрипция"])  # Заголовки

    for word in words:
        transcription = get_transcription(word)
        if transcription:
            writer.writerow([word, transcription])
        else:
            print(f"Не удалось найти транскрипцию для слова {word}")

print("Данные записаны в ipa.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# Функция для извлечения транскрипции с страницы Викисловаря
def get_transcription(word):
    url = f"https://ru.wiktionary.org/wiki/{word}"
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Не удалось получить данные для {word}")
        return None

    soup = BeautifulSoup(response.text, 'html.parser')

    # Ищем транскрипцию
    transcription = None
    for span in soup.find_all("span", {"class": "IPA"}):
        transcription = span.get_text()
        break  # Возьмем только первую найденную транскрипцию

    return transcription

# Чтение слов из файла ru_50k.txt, начиная с 10,001-го слова
words = []
with open('russian_words_all.txt', 'r', encoding='utf-8') as file:
    for i, line in enumerate(file.readlines()):
        if i > 11735:  # Берем только слова после 10,000
            word = line.split()[0]  # Извлекаем только слово
            words.append(word)

# Открываем CSV файл для добавления данных
with open('words.csv', mode='a', newline='', encoding='utf-8') as file:  # Режим "a" для добавления
    writer = csv.writer(file)

    for word in words:
        transcription = get_transcription(word)
        if transcription:
            writer.writerow([word, transcription])
        else:
            print(f"Не удалось найти транскрипцию для слова {word}")

print("Новые данные добавлены в words.csv")
```
